```python
# plot 2 + (n-1)5 = 5n-3

import matplotlib.pyplot as plt
import numpy as np

# Generate x values (n values)
n = np.arange(1, 11)  # Example range from 1 to 10

# Calculate y values (5n - 3)
y = 5 * n - 3

# Plot the line
plt.plot(n, y)
plt.xlabel('n')
plt.ylabel('5n - 3')
plt.title('Plot of 2 + (n-1)5 = 5n - 3')
plt.grid(True)
plt.show()
```
